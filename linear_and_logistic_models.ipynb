```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Load the built-in scikit-learn breast cancer dataset
cancer_data = load_breast_cancer()
X = cancer_data.data
y = cancer_data.target

print(f"Dataset loaded successfully.")
print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
```

```python
# Perform the 80:20 train-test split with the specified random state
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=9001
)

# Initialize and apply MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize Logistic Regression with default solver and fit
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)

# Generate predictions
y_pred = log_reg.predict(X_test_scaled)

# Display the classification report
print("=== Logistic Regression Classification Report ===")
print(classification_report(y_test, y_pred, digits=4))
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error

# Load the built-in scikit-learn diabetes dataset
diabetes_data = load_diabetes()
X_diab = diabetes_data.data
y_diab = diabetes_data.target
feature_names = diabetes_data.feature_names

print(f"Diabetes dataset loaded successfully.")
print(f"Features: {feature_names}")
```

```python
# Perform the 80:20 train-test split with the specified random state
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    X_diab, y_diab, test_size=0.20, random_state=9001
)

# Initialize and apply MinMaxScaler
scaler_d = MinMaxScaler()
X_train_d_scaled = scaler_d.fit_transform(X_train_d)
X_test_d_scaled = scaler_d.transform(X_test_d)

# Define the three configurations requested
models = {
    "Linear Regression (No Regularization)": LinearRegression(),
    "L1 Regularization (Lasso, alpha=1)": Lasso(alpha=1),
    "L2 Regularization (Ridge, alpha=1)": Ridge(alpha=1)
}

# Iterate through models, fit, evaluate, and print coefficients
print("=== Regression Models Evaluation ===\n")
for name, model in models.items():
    # Train
    model.fit(X_train_d_scaled, y_train_d)

    # Predict
    y_pred_d = model.predict(X_test_d_scaled)

    # Calculate Mean Squared Error
    mse = mean_squared_error(y_test_d, y_pred_d)

    print(f"Model: {name}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print("Coefficients:")
    for feat, coef in zip(feature_names, model.coef_):
        print(f"  {feat}: {coef:.4f}")
    print("-" * 50)
```

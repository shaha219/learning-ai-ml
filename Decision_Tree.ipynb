```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report

# Load the breast cancer dataset
cancer_data = load_breast_cancer()
X = cancer_data.data
y = cancer_data.target
feature_names = cancer_data.feature_names
target_names = cancer_data.target_names

print(f"Dataset successfully loaded.")
print(f"Total features: {len(feature_names)}")
```

```python
# 70:30 train:test split, stratified by the target labels, random_state=9001
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=9001, stratify=y
)

# Initialize DecisionTree with max_depth=3 and random_state=9001
dt_classifier = DecisionTreeClassifier(max_depth=3, random_state=9001)

# Fit the tree model on the training set
dt_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = dt_classifier.predict(X_test)

# Print metrics
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred))
print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred, digits=4))
```

```python
# Plot the decision tree structure
plt.figure(figsize=(20, 10))
plot_tree(
    dt_classifier,
    feature_names=feature_names,
    class_names=target_names,
    filled=True,
    rounded=True
)
plt.show()
```
